```python
def tokenize_function(examples):
    articles = [" ".join(article.split()[:max_article_length]) for article in examples["article"]]
    highlight_inputs = [" ".join(highlights.split()[:max_highlight_length]) for highlights in examples["highlights"]]

    tokenized_inputs = tokenizer(articles, padding="max_length", truncation=True, max_length=max_sequence_length)
    tokenized_highlights = tokenizer(highlight_inputs, padding="max_length", truncation=True, max_length=max_sequence_length)

    inputs = {
        "input_ids": tokenized_inputs.input_ids,
        "attention_mask": tokenized_inputs.attention_mask,
        "decoder_input_ids": tokenized_highlights.input_ids,
        "decoder_attention_mask": tokenized_highlights.attention_mask,
    }

    del examples["article"]
    del examples["highlights"]
    del examples["id"]
    return inputs

# Define maximum sequence lengths
max_sequence_length = 150  # 512 --> RAM crash
max_article_length = 150
max_highlight_length = 150

small_tokenized_en_dataset = small_en_dataset.map(tokenize_function, batched=True, batch_size=4)

# Set the format to PyTorch tensors
small_tokenized_en_dataset.set_format("torch")
```

```python
# Set decoding params
model.config.decoder_start_token_id = tokenizer.bos_token_id
model.config.eos_token_id = tokenizer.eos_token_id
model.config.max_length = 150
model.config.min_length = 128
model.config.no_repeat_ngram_size = 3
model.early_stopping = True
model.length_penalty = 2.0
model.num_beams = 12

encoder_length = 150
decoder_length = 128
batch_size = 4
```

```python
# Map data correctly for encoder-decoder inputs

def map_to_encoder_decoder_inputs(batch):
    inputs = {
        "input_ids": batch["input_ids"],
        "attention_mask": batch["attention_mask"],
        "decoder_input_ids": batch["decoder_input_ids"],
        "decoder_attention_mask": batch["decoder_attention_mask"],
    }

    # Check if 'labels' are present
    if "labels" in batch:
        inputs["labels"] = batch["labels"]
    else:
        # Generate 'labels' from 'decoder_input_ids'
        inputs["labels"] = batch["decoder_input_ids"]

    return inputs
```

```python
# Compute metrics for evaluation

def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = tokenizer.eos_token_id
    label_str = tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    rouge_output = rouge.compute(predictions=pred_str, references=label_str, rouge_types=["rouge2"])["rouge2"].mid

    return {
        "rouge2_precision": round(rouge_output.precision, 4),
        "rouge2_recall": round(rouge_output.recall, 4),
        "rouge2_fmeasure": round(rouge_output.fmeasure, 4),
    }
```

```python
# RU
train_dataset_ru = small_tokenized_ru_dataset['train']
train_dataset_ru = train_dataset_ru.map(map_to_encoder_decoder_inputs, batched=True, batch_size=batch_size)
#train_dataset_ru.set_format(type="torch", columns=["input_ids", "attention_mask", "decoder_attention_mask", "decoder_input_ids", "labels"])
train_dataset_ru.set_format(type="torch", columns=["input_ids", "attention_mask", "decoder_attention_mask", "decoder_input_ids", "labels"])

val_dataset_ru = small_tokenized_ru_dataset['validation']
val_dataset_ru = val_dataset_ru.map(map_to_encoder_decoder_inputs, batched=True, batch_size=batch_size)
val_dataset_ru.set_format(type="torch", columns=["input_ids", "attention_mask", "decoder_attention_mask", "decoder_input_ids", "labels"])
```

# **Merge EN & RU datasets**

```python
from torch.utils.data import ConcatDataset, DataLoader, Subset
import random

# Concatenate English and Russian training datasets
merged_train_dataset = ConcatDataset([train_dataset_en, train_dataset_ru])

# Concatenate English and Russian validation datasets
merged_val_dataset = ConcatDataset([val_dataset_en, val_dataset_ru])

# Shuffle the indices of the merged datasets
indices_train = list(range(len(merged_train_dataset)))
indices_val = list(range(len(merged_val_dataset)))
random.seed(24)  #  for reproducibility
random.shuffle(indices_train)
random.shuffle(indices_val)

print("Shuffled Training Indices:", indices_train)              #
print("Shuffled Validation Indices:", indices_val)              #

# Split the shuffled indices into training and validation indices
train_size = len(merged_train_dataset) // 2
val_size = len(merged_val_dataset) // 2

print("Train Size:", train_size)
print("Validation Size:", val_size)


# Create subsets using shuffled indices
subset_train_indices = indices_train[:train_size]
subset_val_indices = indices_val[:val_size]

print("Subset Training Indices:", subset_train_indices)
print("Subset Validation Indices:", subset_val_indices)

# Create DataLoader subsets from the indices
train_dataset = Subset(merged_train_dataset, subset_train_indices)
val_dataset = Subset(merged_val_dataset, subset_val_indices)
```

```python
import logging

from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer

#training_args = TrainingArguments(
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/drive/MyDrive/store_the_checkpoints2", #Google Drive directory! ./
    #output_dir="./",
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    #predict_from_generate=True,
    #evaluate_during_training=True,
    do_train=True,
    do_eval=True,
    logging_steps=375,   #600ex / 4 batch size = 150 make up one epoch !!!
    save_steps=375,        #was 1000
    eval_steps=375,        #was 1000
    overwrite_output_dir=True,
    warmup_steps=2000,
    save_total_limit=3,
    fp16=False,
    save_strategy="steps",  # Save checkpoints based on steps
    learning_rate=5e-5,
    weight_decay=0.01, #TRY 0.5 # prevent overfitting by penalizing large weights in the model.
    predict_with_generate=True,
    num_train_epochs=4, #
)

# Set the pad_token_id attribute of the model's configuration
model.config.pad_token_id = tokenizer.pad_token_id

# instantiate trainer
#trainer = Trainer(
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    compute_metrics=compute_metrics,
    train_dataset=train_dataset,  ## DONT FORGET TO RENAME FOR EN&RU
    eval_dataset=val_dataset, ## DONT FORGET TO RENAME FOR EN&RU
    tokenizer=tokenizer,
    data_collator=data_collator,
)

trainer.train()
```

```python
import torch
from torch.utils.tensorboard import SummaryWriter

# Check if GPU is available and set the device accordingly
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model_inputs = tokenizer(small_en_dataset['test']['article'], padding=True, truncation=True, return_tensors='pt')
model_inputs = {key: tensor.to(device) for key, tensor in model_inputs.items()}  # Move tensors to device

outputs = fine_tuned_model(**model_inputs, output_hidden_states=True)

# Define the directory to save hidden states
path = "/content/drive/MyDrive/results_vis_3-en"
layer=0
if not os.path.exists(path):
    os.mkdir(path)

# Iterate over layers and save hidden states
for layer in range(len(outputs['hidden_states'])):
    layer_path = os.path.join(path, f'layer_{layer}')
    if not os.path.exists(layer_path):
        os.mkdir(layer_path)

    example = 0
    tensors = []
    labels = []

    # Extract hidden states for each example
    for example in range(len(outputs['hidden_states'][layer])):
        sp_token_position = 0
        for token in model_inputs['input_ids'][example]:
            if token != 101:
                sp_token_position += 1
            else:
                tensor = outputs['hidden_states'][layer][example][sp_token_position]
                tensors.append(tensor.to(device))  # Move tensor to device
                break

        label = [small_en_dataset['test']['article'][example], str(small_en_dataset['test']['highlights'][example])] #not tokenized?
        labels.append(label)

    # Write embeddings to TensorBoard
    writer = SummaryWriter(layer_path)
    writer.add_embedding(torch.stack(tensors), metadata=labels, metadata_header=['Text', 'Emotion'])
    layer+=1
```
